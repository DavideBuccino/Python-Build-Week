```python
import pandas as pd

# Carica i due file CSV
df_prodotti = pd.read_csv("Under_Armour_pulito.csv", encoding="utf-8")
df_taglie = pd.read_csv("Under_Armour_Taglie Aggiornato.csv", encoding="utf-8")

# Elimina le righe duplicate dal df_prodotti basandosi su tutte le colonne
df_prodotti_senza_duplicati = df_prodotti.drop_duplicates()

# Elimina le righe duplicate dal df_taglie basandosi su modello, sport e taglie
df_taglie_senza_duplicati = df_taglie.drop_duplicates(subset=['modello', 'sport', 'taglie'])

# Unisci i DataFrame basandosi sulla colonna 'nome_prodotto'
df_unito = pd.merge(df_prodotti_senza_duplicati, df_taglie_senza_duplicati[['nome_prodotto', 'taglie']],
                    on='nome_prodotto',
                    how='left')  # Usa 'left' per mantenere tutti i prodotti di Under Armour

# Visualizza le prime righe del DataFrame unito
print(df_unito.head())
print(f"Numero di righe nel DataFrame unito: {len(df_unito)}")


```
